# Calculate Cross traffic timings for max Interference on each hop

```python
# network
numHops = 3
linkspeed = 100000000 # 100 Mbps
propagationDelay = 0 # 1 microsecond
IFG = 96 # 96 bits
IFGtime = IFG / linkspeed # 960 ns
switchDelay = 0.000008 # 8 microsecond

# app timing
appPayload = 750 # 750 bytes
# appPayload = 1522
appTime = 0.022 # 22 ms transmission start time of the app frame
appTransmissionTime = ((appPayload*8) / linkspeed) + propagationDelay# time to transmit the app frame

# cross traffic
ctPayload = 1526 # 1500 bytes
ctTransmissionTime = ((ctPayload*8) / linkspeed) + propagationDelay # time to transmit the cross traffic frame
```

```python
# calculate worst case cross traffic timing for each hop
crossTrafficTime = 0
BufferTime = IFGtime
for i in range(numHops):
    ctTime = appTime + appTransmissionTime + i*(ctTransmissionTime-IFGtime) - ctTransmissionTime - IFGtime
    print ("Hop", i+1, "cross traffic send time:", ctTime)
    appTransmissionTime = 0
```

```python
# calculate arrival time of packet at the switch
BufferTime = 5*IFGtime
arrivalTime = appTime + appTransmissionTime
# calculate the send time of the cross traffic to meet the packet at the switch
ctSendTime1 = arrivalTime - ctTransmissionTime - IFGtime
ctSendTime2 = ctSendTime1 + ctTransmissionTime - IFGtime - switchDelay
ctSendTime3 = ctSendTime2 + ctTransmissionTime - IFGtime - switchDelay
print("App frame send time: ", appTime)
print("App frame arrival time at switch: ", arrivalTime)
print("Cross traffic frame 1 send time: ", ctSendTime1)
print("Cross traffic frame 2 send time: ", ctSendTime2)
print("Cross traffic frame 3 send time: ", ctSendTime3)
```
